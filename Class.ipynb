```python
from random import *

class Dice:
    def __init__(self, player):
        self.player = player
        
    def run_dice(self):
        self.dice = randint(1, 6)
        print('{0} - 주사위 눈금은 {1}'.format(self.player, self.dice))
        
    def who_win(self, versus):
        if (self.dice > versus.dice):
            print("{0} 승리".format(self.player))
        elif (self.dice < versus.dice):
            print("{0} 승리".format(versus.player))
        else:
            print("비김 {0} vs {1}".format(self.dice, versus.dice))
            
    def get_dice(self):
        return self.dice

player1 = Dice('플레이어1')
player1.run_dice()

player2 = Dice('플레이어2')
player2.run_dice()

player1.who_win(player2)
player2.who_win(player1)
```

```python
#### 1) Game 클래스
#    1-1) 게임을 종료할 상황인지 판정
#    1-2) 비밀 상점 온/오프 판정
#    1-3) 승리자 판정
#    1-4) 비밀 상점 물품 사용 판정
#    1-5) 광역 스킬 시전 판정
# 2) Player 클래스
# 3) Dice 클래스 (2번 문제 코드 재활용)

class Game:
    def __init__(self):
        self.secret_store = False
        self.game_end = False
        self.use_chaos_ticket = False
        self.player_arr = []
        self.dice_sum = 0
        
    def open_secret_store(self, player_arr):
        print("비밀 상점이 열렸으니 플레이어들은 물품을 구매해주세요!!!")
        # TODO: 숙제
        # 비밀 상점 기능은 실제 루프 돌면서 주사위를 돌리는 플레이어들의 행동을 제한해야함
        # 혹은 연속 돌리기 기능의 경우 다른 플레이어들은 정지 상태로 자신만 돌아야함
        
        # 전략
        # 1) 연속 돌리기 - 현재 사용자 외에 모두 정지!
        # 2) 특정 대상의 정지 - 지정한 사용자만 정지!
        
        # 사용자 정보 구성 (사용자의 위치, 랩, 인컴, 이름에 대한 정보가 모두 필요함)
        # Map - key : value
        #       test1: [lap, location, income]
        player_info = {}
        player_length = len(player_arr)
        
        for i in range(player_length):
            player_core_info = []
            player_core_info.append(self.player_arr[i].get_lap())
            player_core_info.append(self.player_arr[i].get_location())
            player_core_info.append(self.player_arr[i].get_money())
            
            player_info[self.player_arr[i].get_player()] = player_core_info
            
        print("키 : 리스트 = ", player_info)
        
        money_gacha = [10000, 20000, 50000, 100000, 200000, 500000, 1000000]
        
        print("player_length = ", player_length)
        
        for i in range(player_length):
            while True:
                not_finish = True
                
                print("현재 사용자 리스트: ", )
                # 최상위 for의 i를 따라가야 하는데
                # 디버깅 메시지로 사용한 i는 항상 끝나면
                # 1에 있으므로 무조건 1에다가만 때려넣게 됨
                for j in range(player_length):
                    print(player_arr[j].__dict__)
                
                print("1. 레드 티켓(5회 정지), 2. 옐로 티켓(2회 정지), 3. 카오스 티켓(랜덤 2회 정지), " +
                      "4. 뽑기(돈), 5. 연속 굴리기(1 ~ 4회), 6. 돈이 없어서 못삽니다.")
                select = int(input("아이템을 고르세요: "))

                if (select == 1):
                    print("레드 티켓 - 정지시킬 대상을 고르세요!!!")

                    while not_finish:
                        idx = int(input("사용자 인덱스를 선택하세요: "))
                        if (idx < player_length and idx >= 0):
                            player_arr[idx].add_stop_count(5)
                            not_finish = False
                            
                    break
                elif (select == 2):
                    print("옐로 티켓 - 정지시킬 대상을 고르세요!!!")
                    
                    while not_finish:
                        idx = int(input("사용자 인덱스를 선택하세요: "))
                        if (idx < player_length and idx >= 0):
                            player_arr[idx].add_stop_count(2)
                            not_finish = False
                            
                    break
                elif (select == 3):
                    print("카오스 티켓 - 랜덤 대상을 정지시킵니다(본인 포함)!!!")
                    
                    while not_finish:
                        idx = randint(0, player_length)
                        if (idx < player_length and idx >= 0):
                            player_arr[idx].add_stop_count(2)
                            not_finish = False
                    
                    break
                elif (select == 4):
                    print("뽑기를 시작합니다!")
                    
                    gacha_idx = randint(0, len(money_gacha) - 1)
                    print("가챠 인덱스 = ", gacha_idx)
                    
                    print("i = {0}, j = {1}".format(i, j))
                    
                    player_arr[i].add_money(money_gacha[gacha_idx])
                    
                    break
                elif (select == 5):
                    print("연속 굴리기(1 ~ 4회)!!!")
                    
                    rand_count = randint(1, 4)
                    print("rand_count = ", rand_count)
                    
                    player_arr[i].set_repeat_count(rand_count)
                    
                    break
                elif (select == 6):
                    print('돈 읎다!')
                    break
                else:
                    print("잘못된 숫자 입력! 다시 선택하세요!")
            
        print("특수 상점 종료!")
        
    def dice_skill(self, player_arr, player_info, cur_dice):
        print("{0} - 주사위 스킬이 발동됩니다!".format(player_info.player))

        # 파이썬에는 switch가 없으므로 if로 구현해야함!
        if cur_dice == 2:
            print("모두의 자금에서 50만원을 깐다.")
            
            for i in range(len(player_arr)):
                player_arr[i].withdraw(500000)
        elif cur_dice == 4:
            print("자신의 자금의 5%가 증발한다.")
            # TODO: 숙제
            player_info.withdraw_percent(5)
        elif cur_dice == 6:
            print("모두의 자금에 30만원을 추가한다.")
            
            for i in range(len(player_arr)):
                player_arr[i].deposit(300000)
        
        
    def start_game(self):
        self.player = int(input("몇 명이 플레이 하는 게임입니꽈 ? "))
        
        for i in range(self.player):
            name = input("플레이어 이름을 적어주세요")
            self.player_arr.append(Player(name))
            self.player_arr[i].repeat_count = 0
            
        while (not(self.game_end)):
            for i in range(self.player):
                # 정지 티켓 계열 및 연속 뽑기 유무 판단이 필요함
                # 누굴 정지 시켜야 하는지
                # 몇 회 정지 해야 하는지
                # ex) A, B - 정지 5회, C
                #     A 굴리고 다음턴 B 정지, C 굴림
                #     A 굴리고 B가 5회 정지하고 C를 굴리면 역시 정지 티켓의 의미가 없음
                # 루프를 돌면서 실제 몇 번 정지했는지 검사하는 루틴이 필요함을 의미함
                repeat_count = self.player_arr[i].repeat_count
                stop_count = self.player_arr[i].stop_count
                
                #tmp_count = self.player_arr[i].repeat_count - self.player_arr[i].stop_count
                tmp_count = repeat_count - stop_count
                
                if (repeat_count == stop_count):
                    self.player_arr[i].repeat_count = 0
                    self.player_arr[i].stop_count = 0
                
                # A가 정지 5회, 연속 4회
                # A가 정지 2회, 연속 2회
                # A가 정지 없음, 연속 2회
                
                while tmp_count >= 0:
                    # A가 정지 5회, 연속 4회 --- 문제 발생(stop_count)
                    # A가 정지 2회, 연속 2회 --- 문제 발생(stop_count)
                    if (self.player_arr[i].stop_count == 0 or tmp_count == 0):
                        self.player_arr[i].run_dice()
                        cur_dice = self.player_arr[i].get_dice()
                        self.dice_sum += cur_dice

                        if (not(cur_dice % 2)):
                            self.dice_skill(self.player_arr, self.player_arr[i], cur_dice)

                        self.player_arr[i].update_location(cur_dice)

                        if (self.player_arr[i].get_lap() == 3):
                            self.game_end = True
                            break
                            
                        # A가 정지 없음, 연속 2회 --- 문제 발생(tmp_count)
                        tmp_count -= 1
                        
                if (self.player_arr[i].stop_count != 0):
                    self.player_arr[i].stop_count -= 1
                    
                self.player_arr[i].repeat_count = 0
                    
            if (self.game_end):
                break
                
            # 실제 비밀 상점이 열려서 동작하는 구간은 이 아래
            if (not(self.dice_sum % 2)):
                self.open_secret_store(self.player_arr)
                
            self.dice_sum = 0
            
        # 정산을 통해 승자 판정(게임을 먼저 끝내도 4번만 나왔으면 질 수도 있음)
        # TODO: 숙제
        # map
        # "이름": 값, "이름": 값
        # 키: 밸류, 키: 밸류
        income = {}
        
        for i in range(len(self.player_arr)):
            print(self.player_arr[i].__dict__)
            
            # 가만보니 이 코드엔 숨겨진 문제가 있음
            # 뭘까요 ?
            # 1) 이 코드의 목적은 무엇인가 ? 결국 정산을 통해 승자를 판정하는 것임
            # 2) 이 코드는 income(수입)이 가장 높은 경우만 찾음(승자를 찾지 못함)
            #income.append(self.palyer_arr[i].get_money())
            
            income[self.player_arr[i].get_player()] = self.player_arr[i].get_money()
            
        print("income: ", income)
        
        # 위의 문제되는 케이스는 Map을 통해 제거했고
        # 추가적으로 value 기반 정렬을 통해 값이 큰 녀석으로 정렬이 가능
        sorted_income = sorted(income.items(), key=value_support, reverse=True)
        
        print("income: ", sorted_income)
        
        print("최종 승자: {0}".format(sorted_income[0]))
        
class Player(Dice):
    def __init__(self, player):
        super(Player, self).__init__(player)
        # 몇 바퀴 돌았는지
        self.lap = 0
        # 현재 위치
        self.location = 0
        # 소지금
        self.money = 0
        # 정지 처분 회수
        self.stop_count = 0
        # 연속 굴리기 회수
        self.repeat_count = 0
        
        print("Player 생성자 호출!")
        
    def update_location(self, dice):
        self.location += dice
        
        if (self.location >= 40):
            self.location -= 40
            self.lap += 1
            
            self.money += 1000000
            
        print("위치값 갱신 lap - {0}, location - {1}".format(self.lap, self.location))
            
    def get_lap(self):
        return self.lap
    
    def get_location(self):
        return self.location
        
    def withdraw(self, money):
        self.money -= money
        print("출금: 내돈 = {0}".format(self.money))
        
    def deposit(self, money):
        self.money += money
        print("입금: 내돈 = {0}".format(self.money))
        
    def withdraw_percent(self, percent):
        #self.money = self.money - self.money * 5 / 100
        #self.money = self.money * (1 - 5 / 100)
        #self.money -= self.money * (5 / 100)
        # a x b = ab
        self.money -= self.money * percent / 100
        print("퍼센트 출금: 내돈 = {0}".format(self.money))
        
    def get_money(self):
        return self.money
    
    def get_player(self):
        return self.player
    
    def add_stop_count(self, count):
        self.stop_count += count
        
    def add_money(self, money):
        self.money += money
        
    def set_repeat_count(self, count):
        self.repeat_count = count
        
board_game = Game()
board_game.start_game()

print(board_game.__dict__)
```
